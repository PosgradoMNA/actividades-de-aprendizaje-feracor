
## 6. Escala los datos, si consideras necesario

```python
X = df.loc[:, df.columns != 'Y'].values 
Y = df['Y'].values 
X_scal = StandardScaler().fit_transform(X) 
escalar_df = pd.DataFrame(X_scal)
escalar_df
```


## 7. Reduce las dimensiones con PCA, si consideras necesario.


```python
df1_pca = PCA(n_components = 10) # Despues de ajustar los componentes a 10, no arroja error
componentes_principales = df1_pca.fit_transform(X_scal)
 
print('Forma antes de PCA: ', X_scal.shape)
print('Forma despues de aplicar el  PCA: ', componentes_principales.shape)
 
df2_pca = pd.DataFrame( data=componentes_principales, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])

target_names = {0.0:'SI',1.0:'NO'}
 
df2_pca['Viable'] = Y 
df2_pca['Viable'] = df2_pca['Viable'].map(target_names) 
 
df2_pca.head()

print(df1_pca.explained_variance_ratio_) 
df1_pca.explained_variance_ratio_.sum() 
df2_pca
```


## 9. Realiza la visualizacin de los datos usando por lo menos 3 grficos que consideres adecuados: plot, scatter, jointplot, boxplot, areaplot, pie chart, pairplot, bar chart, etc.

```python
pca = PCA(n_components=2)
 
columnas_pca = pca.fit_transform(X_scal)
 
pca_df = pd.DataFrame(
    data=columnas_pca, 
    columns=['PC2', 'PC3'])
 
target_names = {
    0.0:'NOT_CREDIBLE_CLIENT',
    1.0:'CREDIBLE_CLIENT'
}
 
pca_df['target'] = Y
pca_df['target'] = pca_df['target'].map(target_names)
 
sns.set()
 
sns.lmplot(
    x='PC2', 
    y='PC3', 
    data=pca_df, 
    hue='target', 
    fit_reg=False, 
    legend=True
    )
 
plt.title('2D PCA Graph')
plt.show()
```
